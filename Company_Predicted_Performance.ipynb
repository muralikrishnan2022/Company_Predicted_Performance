```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
# Load your datasets
train_data = pd.read_csv('/content/drive/MyDrive/train_s.csv')
test_data = pd.read_csv('/content/drive/MyDrive/test_s.csv')
```

```python
# Replace 'Feature_0' and 'Feature_89' with the actual column names
start_column = 'Feature_0'
end_column = 'Feature_89'

# Get the range of columns
feature_columns = train_data.loc[:, start_column:end_column].columns

# Fill null values with the mode of each column
train_data[feature_columns] = train_data[feature_columns].apply(lambda x: x.fillna(x.mode().iloc[0]))
```

```python
train_data.isnull().sum()
```

```python
# Replace 'Feature_0' and 'Feature_89' with the actual column names
start_column = 'Feature_0'
end_column = 'Feature_89'

# Get the range of columns
feature_columns = test_data.loc[:, start_column:end_column].columns

# Fill null values with the mode of each column
test_data[feature_columns] = test_data[feature_columns].apply(lambda x: x.fillna(x.mode().iloc[0]))
```

```python
test_data.isnull().sum()
```

```python
train_data['PERFORMANCE'].unique()
```

```python
# Use LabelEncoder to convert string labels to numerical labels
label_encoder = LabelEncoder()
train_data['PERFORMANCE'] = label_encoder.fit_transform(train_data['PERFORMANCE'])
```

```python
#Display the mapping between original labels and numerical labels
label_mapping = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))
print("Label mapping:", label_mapping)
```

```python
import pandas as pd

# Identify columns with the object data type
object_columns = train_data.select_dtypes(include=['object']).columns

# Remove columns with the object data type
train_data = train_data.drop(object_columns, axis=1)
```

```python
train_data
```

```python
import pandas as pd

# Identify columns with the object data type
object_columns = test_data.select_dtypes(include=['object']).columns

# Remove columns with the object data type
test_data = test_data.drop(object_columns, axis=1)
```

```python
test_data
```

```python
# Import necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from xgboost import XGBClassifier
```

```python
# Split the data into features (X) and target variable (y)
X_train = train_data.drop('PERFORMANCE', axis=1)
y_train = train_data['PERFORMANCE']
```

```python
# Split the data into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)
```

```python
# Initialize XGBoost classifier
model = XGBClassifier()

# Train the model on the training set
model.fit(X_train, y_train)
```

```python
# Make predictions on the validation set
val_predictions = model.predict(X_val)

# Evaluate the accuracy of the model on the validation set
accuracy = accuracy_score(y_val, val_predictions)
print(f"Accuracy on validation set: {accuracy:.2f}")
```

```python
# predictions on the test data
test_predictions = model.predict(test_data)

# Create a new DataFrame to store the results
result_df = pd.DataFrame({
    'id': test_data['id'],
    #'Predicted_Performance': test_predictions,
    'PERFORMANCE': [next(key for key, value in label_mapping.items() if value == pred) for pred in test_predictions]
})
```

```python
# Save the results to a new CSV file
result_df.to_csv('/content/drive/MyDrive/predicted_performance.csv', index=False)
```
